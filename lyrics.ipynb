```python
data_aus = pd.read_csv("Australia.csv", header=1)
data_brz = pd.read_csv("Brazil.csv", header=1)
data_can = pd.read_csv("Canada.csv", header=1)
data_col = pd.read_csv("Colombia.csv", header=1)
data_cor = pd.read_csv("Costa_Rica.csv", header=1)
data_fra = pd.read_csv("France.csv", header=1)
data_ger = pd.read_csv("Germany.csv", header=1)
data_gua = pd.read_csv("Guatemala.csv", header=1)
data_hk = pd.read_csv("HK.csv", header=1)
data_ind = pd.read_csv("India.csv", header=1)
data_jap = pd.read_csv("Japan.csv", header=1)
data_mex = pd.read_csv("Mexico.csv", header=1)
data_neth = pd.read_csv("Netherlands.csv", header=1)
data_nz = pd.read_csv("NZ.csv", header=1)
data_phi = pd.read_csv("Philipines.csv", header=1)
data_sng = pd.read_csv("Singapore.csv", header=1)
data_swe = pd.read_csv("Sweden.csv", header=1)
data_tur = pd.read_csv("Turkey.csv", header=1)
data_uk = pd.read_csv("UK.csv", header=1)
data_us = pd.read_csv("US.csv", header=1)
```

```python
list_of_data_files = [data_aus, data_brz, data_can, data_col, data_cor, data_fra, data_ger, 
                      data_gua, data_hk, data_ind, data_jap, data_mex, data_neth, data_nz, 
                      data_phi, data_phi, data_sng, data_swe, data_tur, data_uk, data_us]
list_of_text_files = ["Australia_lyrics.txt", "Brazil_lyrics.txt", "Canada_lyrics.txt", "Colombia_lyrics.txt", "Costa_Rica_lyrics.txt", "France_lyrics.txt", "Germany_lyrics.txt", "Guatemala_lyrics.txt", "Hong_Kong_lyrics.txt",
            "India_lyrics.txt", "Japan_lyrics.txt", "Mexico_lyrics.txt", "Netherlands_lyrics.txt", "New_Zealand_lyrics.txt", "Philippines_lyrics.txt", "Singapore_lyrics.txt", "Sweden_lyrics.txt", "Turkey_lyrics.txt", "United_Kingdom_lyrics.txt", "United_States_lyrics.txt"];

for data_file_index in range(len(list_of_data_files)):
    data_file = list_of_data_files[data_file_index]
    text_file = list_of_text_files[data_file_index]
    f = open(text_file, "w+")
    counter = 0
    for index, row in data_file.iterrows():
        if counter >= 50:
            break
        #if (row['Track Name'] === "Outta My Head (with John Mayer)"):
        #    row['Track Name'] = "Outta My Head"
            
        if ("(with" in row['Track Name']):
            #print(row['Track Name'])
            row['Track Name'] = row['Track Name'].split("(with",1)[0] 
        song = genius.search_song(row['Track Name'], row['Artist'])
        #print(type(song))
        if (song != None):
            if(detect(song.lyrics) == 'en'):
                #print(song.lyrics)
                #append to some document
                f.write(song.lyrics + "\n")
        counter = counter + 1
    f.close()
```
